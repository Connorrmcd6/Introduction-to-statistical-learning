```python
def euclidean(x1, x2):
    d = abs(x1 - x2)
    return d
```

```python
def neighbors(v , data, k):
    x1 = v
    vals = []
    dis = []
    r = 0
    
    for i in data:
        d = euclidean(x1, i)
        vals.append([r, d])
        dis.append(d)
        r += 1
    dis = np.sort(dis)
    returns = []
    pos = []    
    for i in range(0, k):
        for j in vals:
            if j[1] == dis[i]:
                returns.append(j)
                pos.append(j[0])
                
    return pos[-1]
```
